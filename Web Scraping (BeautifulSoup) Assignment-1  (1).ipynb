# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’.


We have to send request to the web page and get response 200 to scrap the required data from the web page.

```python
page=requests.get("https://bookpage.com/reviews")
page
```

Since the response is 200 we can scrap data from this website.

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
url=soup.find_all("h4",class_="italic")
url
```

```python
urls=[]
for i in url:
    for j in i.find_all("a",href=True):
        urls.append(j['href'])
urls
```

Extracting book name.

```python
book_name=[]
for i in urls:
    name=requests.get('https://bookpage.com'+i)
    soup=BeautifulSoup(name.content,'html.parser')
    nam=soup.find("h4",class_="italic")
    book_name.append(nam.text.replace('\n','').replace('★',''))
book_name
```

```python
book_name=book_name[:5]
```

```python
len(book_name)
```

Extracting author name.

```python
author_name=[]
for i in urls:
    auna=requests.get('https://bookpage.com'+i)
    soup=BeautifulSoup(auna.content,'html.parser')
    aut=soup.find("p",class_="sans bold")
    author_name.append(aut.text.replace('\n',''))
author_name
```

```python
author_name=author_name[:5]
```

```python
len(author_name)
```

Extracting reviews.

```python
reviews=[]
for i in urls:
    book=requests.get('https://bookpage.com'+i)
    soup=BeautifulSoup(book.content,'html.parser')
    rev=soup.find("div",class_="article-body")
    reviews.append(rev.text)
reviews
```

```python
reviews=reviews[:5]
```

```python
len(reviews)
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
soup=BeautifulSoup(page.content)
```

Extracting genre.

```python
gen=soup.find_all("p",class_="genre-links hidden-phone")
gen
```

```python
genre=[]
for i in gen:
    for j in i.find_all("a"):
        genre.append(j.text.replace("\n"," "))
genre
```

```python
genre=genre[:5]
```

```python
len(genre)
```

```python
print(len(book_name),len(author_name),len(reviews),len(genre))
```

# Frameing the dataset:

Importing pandas and Frameing data.

```python
import pandas as pd
book_page=pd.DataFrame({})
book_page['Name']=book_name
book_page['Author']=author_name
book_page['Reviews']=reviews
book_page['Genre']=genre
```

```python
book_page
```

This is the required DataFrame.
